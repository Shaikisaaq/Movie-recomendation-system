```python
import pandas as pd 
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')

print(movies.shape,"  ", credits.shape)
```

```python
movies.head(10)['title']
```

```python
credits.head(10)['title']
```

```python
# merging the both dataframe based on titled

movies=movies.merge(credits,on='title')
```

```python
# fitering of dataset by extracting only important colums

movies=movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head(5)
```

```python
# checking whether the data has null values if then then delete or drop it

movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
movies.isnull().sum()
```

```python
# checking weather the dataset has any duplicate values if there then delete or drop
movies.duplicated().sum()
```

```python
# now preprocessing on cloumn who are multivalued
import ast

def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l

movies['genres']=movies['genres'].apply(convert)
movies['keywords']=movies['keywords'].apply(convert)
movies
        
```

```python
# extracting top 3 cast names from the cast column

def convert(obj):
    ls=[]
    count=0
    for i in ast.literal_eval(obj):
        if count !=3:
            ls.append(i['name'])
            count+=1
        else:
            break
    return ls

movies['cast']=movies['cast'].apply(convert)
```

```python
movies.head()
```

```python
# extacting only director names from the crew column
def director(obj):
    l=[]
    for i in ast.literal_eval(obj):
        if i['job']== 'Director':
            l.append(i['name'])
            
    return l
```

```python
movies['crew']=movies['crew'].apply(director)

movies.head(3)
```

```python
# converting overview into a list 
movies['overview']=movies['overview'].apply(lambda x:x.split())
movies.head(3)
```

```python
# removing unessary space btw names because it will be difcuilty to create tag for that names 
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['overview']=movies['overview'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])

movies.head(3)
```

```python
# create a new column called as tag by combine all the list

movies['tag']= movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
table=movies[['movie_id','title','tag']]
table.head(3)
```

```python
# convering the list into string 
table['tag']=table['tag'].apply(lambda x: " ".join(x))
table.head(3)
```

```python
table['tag']=table['tag'].apply(lambda x: x.lower() )
table.head(3)
```

```python
table.iloc[0].tag
```
